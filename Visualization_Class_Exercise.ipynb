```r
library(ggplot2)
data <- ggplot2::mpg
data$Transmission <- ifelse(startsWith(data$trans, 'a'), "automatic", "manual")
plot <- ggplot(data, aes(x = cty, y = hwy)) + 
facet_grid(Transmission ~ year) +
ggtitle("Fuel Economy Data from 1999 to 2008") + xlab("City Miles per Gallon") + ylab("Highway Miles per Gallon") + 
theme(axis.title.x = element_text(color="green", size=15, face="bold", margin = margin(t = 25, r = 0, b = 0, l = 0)),
      axis.title.y = element_text(color="blue", size=15, face="bold", margin = margin(t=0, r=30, b=0, l=0)),
      plot.title= element_text(color="green", size=20, face="bold", hjust=0.5),
      strip.background= element_rect(color="red", fill="blue")) +
      geom_col(fill="yellow", color="blue")
plot
```
